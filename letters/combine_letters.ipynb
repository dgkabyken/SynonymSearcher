```python
import pandas as pd
import numpy as np
import pickle
import collections
pd.options.display.max_rows = 10
```

```python
file = open('А.pkl', 'rb')
d1 = pickle.load(file)
file.close()
file = open('Б.pkl', 'rb')
d2 = pickle.load(file)
file.close()
file = open('В.pkl', 'rb')
d3 = pickle.load(file)
file.close()
file = open('Г.pkl', 'rb')
d4 = pickle.load(file)
file.close()
file = open('Д.pkl', 'rb')
d5 = pickle.load(file)
file.close()
file = open('ЕЁЖЗ.pkl', 'rb')
d6 = pickle.load(file)
file.close()
file = open('И.pkl', 'rb')
d7 = pickle.load(file)
file.close()
file = open('Й.pkl', 'rb')
d8 = pickle.load(file)
file.close()
file = open('КЛМНОП.pkl', 'rb')
d9 = pickle.load(file)
file.close()
file = open('Р.pkl', 'rb')
d10 = pickle.load(file)
file.close()
file = open('СТУФХШЩЫЭЮЯ.pkl', 'rb')
d11 = pickle.load(file)
file.close()
file = open('Ц.pkl', 'rb')
d12 = pickle.load(file)
file.close()
file = open('Ч.pkl', 'rb')
d13 = pickle.load(file)
file.close()
```

```python
d = {} # словарь из синонимов для всех букв
all_letters = [d1, d2, d3, d4, d5, d6, d7, d8, d9, d10, d11, d12, d13]
for letter in all_letters:
    d.update(letter)
```

```python
ordered_d = collections.OrderedDict(sorted(d.items())) # отсортированный словарь для всех букв
```

```python
# origin - таблица из слов и их id
origin = pd.DataFrame({'word': list(ordered_d.keys()), 'id': np.arange(1, len(ordered_d.keys()) + 1)})
origin
```

```python
syn = [] # все синонимы
tmp = []
id_ = [] # id соответствующие этим синонимам
for key in ordered_d.keys():
    syn.extend(ordered_d[key])
    tmp.append(len(ordered_d[key]))
    
for i, each in enumerate(tmp):
    id_.extend([i + 1]*each)
    
del tmp
```

```python
# synonym - таблица из синонимов и их id: внутренние и внешние(id из таблицы origin)
synonym = pd.DataFrame({'id': id_, 'id_inner': np.arange(1, len(id_) + 1), 'syn': syn})
synonym
```

```python
# save tables in csv
origin.to_csv('origin.csv', index = False)
synonym.to_csv('synonym.csv', index = False)
```
