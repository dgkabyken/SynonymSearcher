```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
import time
import string
from random import randint
from fake_useragent import UserAgent
import pickle
```

```python
a = ord('А')
alphabet = [chr(i) for i in range(a, a + 6)] + [chr(1025)] + [chr(i) for i in range(a + 6, a + 32)]
```

```python
def get_meta_by_letter(letter):
    url = 'https://synonymonline.ru/{}'.format(letter)
    r = requests.get(url, headers = {'User-Agent': UserAgent().chrome})
    soup = BeautifulSoup(r.text)
    pages = soup.find_all('ul', {'class': 'pagination'})
    if len(pages) == 0:
        last_page = 1
    else:
        last_page = int(pages[0].find_all('a')[-1].get('href')[8:])
    return soup, last_page
```

```python
def get_syn_urls(page, soup):
    if page == 0:
        tables = soup.find_all('ul', {'class': 'list-unstyled words-list'})
        for item in tables:
            hrefs = item.find_all('a')
            for each in hrefs:
                urls[letter].append('https://synonymonline.ru' + each.get('href'))
       
    else:
        for i in range(5):
            try:
                url = 'https://synonymonline.ru/{}?page={}'.format(letter, page + 1)
                r = requests.get(url, headers = {'User-Agent': UserAgent().chrome})
                soup = BeautifulSoup(r.text)
                tables = soup.find_all('ul', {'class': 'list-unstyled words-list'})
                for item in tables:
                    hrefs = item.find_all('a')
                    for each in hrefs:
                        urls[letter].append('https://synonymonline.ru' + each.get('href'))
                break
            except:
                print(letter, page, 'попытка 1')
                continue
```

```python
def get_urls_by_letter(letter):
    soup, last_page = get_meta_by_letter(letter)
    urls[letter] = []
    for i in range(last_page):
        get_syn_urls(i, soup)
        time.sleep(randint(1, 4))
        #print(letter, i)
```

```python
%%time
urls = {}
for letter in alphabet:
    print(letter, 'я тут!')
    get_urls_by_letter(letter)
    time.sleep(1)
```
